```python
def initializeboard():
    board = np.zeros((8,8))
    board[3,4] = 1
    board[4,3] = 1
    board[3,3] = 2
    board[4,4] = 2
    return board
```

```python
def rules(board, move_y, move_x, player):
    new_board = copy.deepcopy(board)
    new_board[move_y, move_x] = player
    
    
    right = move_x + 1 
    for i in range(right,8):
        if new_board[move_y,right] == 0:
            break
        elif new_board[move_y,right] == player:
            new_board[move_y, move_x:right] = player
            break 
        right += 1
        
        
    left = move_x - 1 
    for i in range(0,left):
        if new_board[move_y,left] == 0:
            break
        elif new_board[move_y,left] == player:
            new_board[move_y, left:move_x] = player
            break
        left -= 1
        
    up = move_y + 1
    for i in range(up,8):
        if new_board[up,move_x] == 0:
            break
        elif new_board[up,move_x] == player:
            new_board[move_y:up, move_x] = player
            break
        up += 1
        
        
    down = move_y - 1
    for i in range(0,down):
        if new_board[down,move_x] == 0:
            break
        elif new_board[down,move_x] == player:
            new_board[down:move_y, move_x] = player
            break
        down -= 1
        
    diag_left_up_1 = move_y -1 
    diag_left_up_2 = move_x -1 
    
    for i in range(0,min(diag_left_up_1,diag_left_up_2)):
        if new_board[diag_left_up_1,diag_left_up_2] == 0:
            break
        elif new_board[diag_left_up_1,diag_left_up_2] == player:
            a = 0
            for i in range(0,move_y - diag_left_up_1):
                new_board[move_y-a, move_x-a] = player
                a += 1
            break
        diag_left_up_1 -= 1
        diag_left_up_2 -= 1
        
        
    diag_right_down_1 = move_y + 1 
    diag_right_down_2 = move_x + 1 
        
    for i in range(max(diag_right_down_1,diag_right_down_2),8):
        if new_board[diag_right_down_1,diag_right_down_2] == 0:
            break
        elif new_board[diag_right_down_1,diag_right_down_2] == player:
            a = 0
            for i in range(0,diag_right_down_1-move_y):
                new_board[move_y+a, move_x+a] = player
                a += 1
            break
        diag_right_down_1 += 1
        diag_right_down_2 += 1
        
    diag_left_down_1 = move_y + 1 
    diag_left_down_2 = move_x - 1 
        
    for i in range(0,min(8-diag_left_down_1,diag_left_down_2)):
        if new_board[diag_left_down_1,diag_left_down_2] == 0:
            break
        elif new_board[diag_left_down_1,diag_left_down_2] == player:
            a = 0
            for i in range(0,diag_left_down_1-move_y):
                new_board[move_y+a, move_x-a] = player
                a += 1
            break
        diag_left_down_1 += 1
        diag_left_down_2 -= 1
        
    diag_right_up_1 = move_y - 1 
    diag_right_up_2 = move_x + 1 
    
        
    for i in range(0,min(diag_right_up_1,8-diag_right_up_2)):
        if new_board[diag_right_up_1,diag_right_up_2] == 0:
            break
        elif new_board[diag_right_up_1,diag_right_up_2] == player:
            a = 0
            for i in range(0,move_y-diag_right_up_1):
                new_board[move_y-a, move_x+a] = player
                a += 1
            break
        diag_right_up_1 -= 1
        diag_right_up_2 += 1
    
    
    return new_board
```

```python
def check_moves(board,player):
    allowed_moves=[]
    for i in range(0,8):
        for j in range(0,8):
            if board[i,j] == 0:
                right = j + 1 
                for k in range(0,8-right):
                    if board[i,right] == 0:
                        break
                    elif board[i,right] == player and k == 0:
                        break
                    elif board[i,right] == player and k > 0:
                        allowed_moves.append([i,j])        #fixa så att större loopen breakar
                        break
                    right += 1
                    
                left = j - 1    
                    
                for k in range(0,left + 1):
                    if board[i,left] == 0:
                        break
                    elif board[i,left] == player and k == 0:
                        break
                    elif board[i,left] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    left -= 1
                 
                up = i - 1
                # range_up = i - 1
                
                for k in range(0,up + 1):
                    if board[up,j] == 0:
                        break
                    elif board[up,j] == player and k == 0:
                        break
                    elif board[up,j] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    up -= 1
                    
                down = i + 1
                    
                for k in range(0,8-down):
                    if board[down,j] == 0:
                        break
                    elif board[down,j] == player and k == 0:
                        break
                    elif board[down,j] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    down += 1
                    
                    
                
                diag_left_up_1 = i - 1 
                diag_left_up_2 = j - 1
                
                for k in range(0,min(diag_left_up_1 + 1,diag_left_up_2 + 1)):
                    if board[diag_left_up_1,diag_left_up_2] == 0:
                        break
                    elif board[diag_left_up_1,diag_left_up_2] == player and k == 0:
                        break
                    elif board[diag_left_up_1,diag_left_up_2] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    diag_left_up_1 -= 1 
                    diag_left_up_2 -= 1
                    
                    
                diag_right_down_1 = i + 1 
                diag_right_down_2 = j + 1
            
                for k in range(0,min(8 - diag_right_down_1,8 - diag_right_down_2)):
                    if board[diag_right_down_1,diag_right_down_2] == 0:
                        break
                    elif board[diag_right_down_1,diag_right_down_2] == player and k == 0:
                        break
                    elif board[diag_right_down_1,diag_right_down_2] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    diag_right_down_1 += 1 
                    diag_right_down_2 += 1
                
                
                diag_right_up_1 = i - 1
                diag_right_up_2 = j + 1
                
                for k in range(0,min(diag_right_up_1 + 1,8 - diag_right_up_2)):
                    if board[diag_right_up_1,diag_right_up_2] == 0:
                        break
                    elif board[diag_right_up_1,diag_right_up_2] == player and k == 0:
                        break
                    elif board[diag_right_up_1,diag_right_up_2] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    diag_right_up_1 -= 1 
                    diag_right_up_2 += 1
                    
                    
                diag_left_down_1 = i + 1
                diag_left_down_2 = j - 1    
                
                for k in range(0,min(8-diag_left_down_1,diag_left_down_2 + 1)):
                    if board[diag_left_down_1,diag_left_down_2] == 0:
                        break
                    elif board[diag_left_down_1,diag_left_down_2] == player and k == 0:
                        break
                    elif board[diag_left_down_1,diag_left_down_2] == player and k > 0:
                        allowed_moves.append([i,j])
                        break
                    diag_left_down_1 += 1 
                    diag_left_down_2 -= 1
                    
                
                
    return allowed_moves
        
```

```python
def symmetry_check(board_1, board_2):
    if ((board_1 == np.rot90(board_2)).all() or (board_1 == np.rot90(np.rot90(board_2))).all() 
        or (board_2 == np.rot90(board_1)).all() or (board_1 == board_2).all()):
        return True
    
    mirror_board = mirror(board_1, "x")
    
    if ((mirror_board == np.rot90(board_2)).all() or (mirror_board == np.rot90(np.rot90(board_2))).all() 
        or (board_2 == np.rot90(mirror_board)).all() or (mirror_board == board_2).all()):
        return True
        
    else:
        return False
```

```python
def mirror(board,x_or_y):         ### if not used anywhere but symmetry_check remove x_y option
    new_board = np.zeros((8,8))
    if x_or_y == "y":
        for i in range(0,8):
            for j in range(0,8):
                new_board[i,j] = board[7-i,j]
    if x_or_y == "x":
        for i in range(0,8):
            for j in range(0,8):
                new_board[i,j] = board[i,7-j]
    return new_board
        
```

```python
def check_history(board, library, player):     # ha ett separat library för varje "steg"
    for i in range(0,len(library)):
        if symmetry_check(board, library[i].board) and library[i].player != player:
            return [True, library[i]]
    return [False,1]
```

```python
def check_actual_choices(board, player):               # delete final_board_list
    allowed_moves = check_moves(board, player)
    board_list = []
    for i in range(0,len(allowed_moves)):
        board_list.append(rules(board,allowed_moves[i][0], allowed_moves[i][1], player))
    #final_board_list = copy.deepcopy(board_list)    
    
    
    remove_list = []
    
    for i in range(0,len(board_list)):
        if not len(remove_list) == len(board_list) - 1:
            for j in range(i+1,len(board_list)):
                if symmetry_check(board_list[i], board_list[j]):
                    remove_list.append(i)
                    break
        else:
            break
    
    #print(remove_list)
    
    for index in sorted(remove_list, reverse=True):
        #del final_board_list[index]
        del allowed_moves[index]
        
    
    return allowed_moves
            
```

```python
library = []

class othello_node():         # lägg twice om action är tom
    def __init__(self, board, parent = None, root = False):
        self.board = board
        self.children = []
        self.parents = []
        if parent != None:
            self.parents.append(parent)
        self.visits = 1;      #N
        self.reward = 0;      #Q
        #self.marker = 1
        self.player = 1
        self.root = root
        self.iter_count = 0
        self.nr_of_choices = check_actual_choices(self.board, self.player)
        if not root:
            if parent.player == 1:
                self.player = 2
        self.actions = check_actual_choices(self.board, self.player)
      

    def add_child(self):
            """Method to create a child node to this parent node - in other words play randomly"""
            new_board = copy.deepcopy(self.board)  #copy current board
            if len(self.actions) > 0:                
                move = random.randint(0, len(self.actions) - 1)              #dubbelkolla -1
                new_board = rules(new_board,self.actions[move][0], self.actions[move][1], self.player)
                self.actions.pop(move)         
            
            played_before = check_history(new_board, library, self.player)
            
            #if not played_before[0]:
            #    self.children.append(othello_node(new_board, parent=self)) #create child and adds to child list
            #    library.append(self.children[-1])      #add child to library
            #else:
            #    self.children.append(played_before[1])                    #add child to childlist
            #    
            
            
            if not played_before[0]:
                self.children.append(othello_node(new_board, parent=self)) #create child
                library.append(self.children[-1])
            else: 
                print(played_before[1].board)
                self.children.append(played_before[1]) #create child
                current_child = self.children[-1]
                current_child.parents.append(self)  #add self to childs parents
                #library.append(self.children[-1])
                           
    def empty_board_check(self):
        """Checking if board is empty"""
        
        for i in range(0,8):
            for j in range(0,8):
                if self.board[i,j] == 0:
                    return False
        return True

    def endgame_check(self):
        no_choice = 0
        end = False
        if self.empty_board_check():
            end = True
        elif len(self.parents) != 0:              ### oeffektivt   behövs?
            if len(self.nr_of_choices) == 0: 
                no_choice += 1
            else:
                no_choice = 0
            if no_choice == 2:
                end = True
        if end:
            self.win_check()
            print(end)
            return True
        return False

    def win_check(self):
        one_count = 0
        two_count = 0
        for i in range(0,8):
            for j in range(0,8):
                if self.board[i,j] == 1:
                    one_count += 1
                elif self.board[i,j] == 2:
                    two_count += 1
      
        if one_count > two_count:
            result = 1
        elif two_count > one_count:
            result = -1
        else:
            result = 0

        self.backpropagate(result)


    def backpropagate(self, result): #result is 1 for X winning, 0 for draw, -1 for O winning
        """Method for backpropagation to the root (node without parent). Recursive."""
        self.visits += 1        # ändra kanske


        if not self.root: 
            #self.parent.visits += 1
            #self.parent.reward += result
            #self.visits += 1
            self.reward += result
            #print(self.parents)
            for parent in self.parents:
                #print('-----')
                #    #print(type(parent))
                #print(self.parents[0].board)
                parent.backpropagate(result)
            #self.parents[0].backpropagate(result)

    def explore_to_leaf(self):
        """Recursive roll-out method for when there are unexplored children"""
        result = self.endgame_check()
        if not result:
            self.add_child()          
            child = self.children[len(self.children)-1]
            child.explore_to_leaf()


    def choose_child(self):
        """Walks down the tree accoring to UCB or backpropagates to root"""
        if not self.endgame_check():
            if len(self.actions) > 0:        # if there are unexplored children
                self.explore_to_leaf()            
            else:
                UCB_max = -float('inf') #for safety margins in loosing trajectories
                UCB_min = float('inf') 
                for child in self.children:
                    
                    if self.player == 2:         #if simulating an O now, minimize the UCB instead
                        UCB = child.reward/child.visits + -2*math.sqrt(np.log(self.visits)/child.visits)  # save in class?
                        if UCB < UCB_min:  
                            UCB_min = UCB
                            best_child = child
                            
                    else:
                        UCB = child.reward/child.visits + 2*math.sqrt(np.log(self.visits)/child.visits)  # save in class?
                        if UCB > UCB_max: 
                            UCB_max = UCB
                            best_child = child
                                
                        #    ----  försök slut  -----        
                        
                #print(self.print_board())           # HÄR ÄR VI
                
        
                if not best_child.endgame_check():
                    best_child.choose_child()
    

    def make_move_both(self):
        """ Chooses the best move to make for X
            return: move: the tic_tac_Node of the best child, e.i. the new state in the tree. 
                    Note that rewards and visits stay. 
        """
        iter = 100
        best_prob = -2
        for i in range(0,iter):            
            self.choose_child()
            print('hi')
            print(self.children[0].reward)
            self.iter_count += 1
            print(self.iter_count)
        

        
        #print ("all the children: ", len(self.children))
        for child in self.children:    #check which child has the highest score
            score=child.reward/(child.visits + 0.0000000001) #to avoid dicision by 0      - tror detta är lugnt/O
            if not self.player == 2:    #if simulating an O now we want to minimize
                score = -score
            #print("No visits {}, reward there {}, score ratio: {}".format(child.visits, child.reward, score))
            if score >= best_prob:
                best_child = child
                best_prob = score
                                                     # detta va en quickfix
        print(f'The score from this move is: {score}')
        
        #print('The move we make has score {}, and is:'.format(best_prob))
        #print(best_child.print_board_new())
        move = best_child
        return move
      
    
                
          
```
